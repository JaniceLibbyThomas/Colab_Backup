```python
n= int(input())
res=""
for i in range(0,n):
  if(i>1):
    for j in range(2,n):
      if(i%j)==0:
        break
    else: 
      res += i+" "
    
print(res.rstrip())
```

```python
# Python program to display all the prime numbers within an interval

n = int(input())
res=""

#print("Prime numbers between", lower, "and", upper, "are:")

for num in range(1, n +1):
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           res+=str(num)+" "
print(res.rstrip())
```

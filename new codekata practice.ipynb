```python
s = input()
lst=list(map(str, s))

for i in range (0,len(lst)):
    for j in range (i+1,len(lst)):
      #print("value of i and j: ",lst[i],lst[j])
      if(lst[i].lower()==lst[j]):
        lst[i]=lst[i].upper()
        lst[j]=lst[j].upper()

for i in range (0,len(lst)):
  print(lst[i],end="")
```

#BAR CHART

```python
#BAR CHART PLOTTING
 
import matplotlib.pyplot as plt
Language = ('Python', 'C++', 'Java', 'Perl', 'Scala')
Useage = [10,8,6,4,2]
plt.figure(facecolor='#FFFAF0')
 #
ax=plt.axes()
ax.set_facecolor("#FFDEDE")

plt.bar(Language, Useage, align='center', alpha= 1,color="#978ABF",width=0.5,hatch=".")
 
#alpha -->transparency of the bar 0 -1 (0.1,0.2....)
#color --> color of the bar
#width --> width of each bars 0,0.1,0.2.....1
#hatch --> style in the bar (+,-,_,^,.,*....)

#plt.barh(performance,width=1) #BARS IN HORIZONTAL
#Matplotlib allows you to regulate the transparency of a graph plot using the alpha attribute. By default, alpha=1
plt.ylabel("USAGE") #y axis name
plt.xlabel("LANGUAGE")
plt.legend(['LANGUAGE'],loc=1)
plt.title("programming language usage") #represents title
plt.show()
```

```python
#STACKED BAR PLOT

from matplotlib import pyplot as plt 
import numpy as np 
countries = ['USA', 'India', 'China', 'Russia', 'Germany'] 
bronzes = np.array([38, 17, 26, 19, 15]) 
silvers = np.array([37, 23, 18, 18, 10]) 
golds = np.array([46, 27, 26, 19, 17]) 
ind = [x for x, _ in enumerate(countries)] 
plt.bar(ind, golds, width=0.5, label='golds', color='gold', bottom=silvers+bronzes) 
plt.bar(ind, silvers, width=0.5, label='silvers', color='silver', bottom=bronzes) 
plt.bar(ind, bronzes, width=0.5, label='bronzes', color='#CD853F') 
plt.xticks(ind, countries) 
plt.ylabel("Medals") 
plt.xlabel("Countries") 
plt.legend(loc=1) 
plt.title("2019 Olympics Top Scorers")
```

```python

import matplotlib.pyplot as plt
import numpy as np
 
city=['Delhi','Beijing','Washington','Tokyo','Moscow']
Gender=['Male','Female']
pos = np.arange(len(city))
Happiness_Index_Male=[60,40,70,65,85]
Happiness_Index_Female=[30,60,70,55,75]
 
plt.bar(pos,Happiness_Index_Male,color='blue',edgecolor='black')
plt.bar(pos,Happiness_Index_Female,color='pink',edgecolor='black',bottom=Happiness_Index_Male)
plt.xticks(pos, city)
plt.xlabel('City', fontsize=16)
plt.ylabel('Happiness_Index', fontsize=16)
plt.title('Stacked Barchart - Happiness index across cities',fontsize=18)
plt.legend(Gender,loc=2)
plt.show()
```

#BOX PLOT

BOX PLOT CAN BE USED 
1. COMPARISON
2. OUTLIERS - EXTREME/ NOISY DATA
3. AREAS WHERE WE LIKE TO GET MAX MIN,MEAN

ASSUMPTION:
TOP- MAX
BOTTOM - MIN
BOX - MEAN

HOW BOX PLOTS

IQR BASED

INTER QUATILE RANGE: 
Consider 40-75 as  range

split into 4 Quater total -35 value split by 35/4 

find median ->Q1
find median -> Q2

plot is Q3 --> most frequent area

```python
import pandas as pd
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/chico.csv"
df=pd.read_csv(url)
print(df.head(5))
 
fig=plt.figure()
test1 = fig.add_subplot(1,2,1)#nrows, ncols, index
test2 = fig.add_subplot(1,2,2)
#pos is a three-digit integer where the first, second, and third integer are nrows,ncols, index.
test1.boxplot(df["grade_test1"],showfliers=False)#method1   #showfliers=False(Hide outliers)
test2.boxplot(df["grade_test2"],showfliers=True)
 
plot = df[['grade_test1', 'grade_test2']].plot(kind='box', title='boxplot', showmeans=True)#method2 1-PIC NO SEPERATION
 
plt.show()
```

#SCATTER PLOT

SCATTER PLOT:

-DISTRIBUTED DOTS BASED UPON DATA

BUBBLE PLOT:

-EXPANDING THE DOTS BASED ON SCENARIO

-COVID19INDIA.ORG

```python
#colormap sample
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
 
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
           cmap='viridis') #CUS OF SIZE IT BECOME BUBBLE PLOT
plt.colorbar();  # show color scale
```

```python
#plotting from iris dataset
from sklearn.datasets import load_iris
iris = load_iris() #INBUILD DATASET AVAILABLE 
features = iris.data.T
#print(features)
 
plt.scatter(features[0], features[1], alpha=0.2,
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

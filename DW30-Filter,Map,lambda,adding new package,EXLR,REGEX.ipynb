1. **FILTER** : INBUILD FUNCTION IN PYTHON


*   Returns an iterator where items are filtered throught a function defined to test if item is accepted or not

*   if items are passed that respective items will be stored in the iterator it could be list, tuple,dict or set.

*  **Syntax: 

```
# This is formatted as code
```

filter(funtion, iterator)**


````python
# Example of filter

def even_checker(x):
  if(x%2==0):
    return True
  else:
    return False

#calling the funtion
list1=[1,2,3,4,5]
filtered_list= list(filter(even_checker,list1))
````

````python
filtered_list # calling filtered list returns a list
````

````python
# to print a filter use for loop
# NOTE: values are seen only once after doing filter if user try to execute the loop again the value wont be displayed
#in order to overcome it save the filtered value as list and print list.
for i in filtered_list:
  print (i)
````

**MAP :**


1.   same as filter needs a function and iterator the only change is it returns some value where in filter we will get the boolean as return type.
2.   **Syntax:map(function,iterator)**



````python
def sqr_num(x):
  return(x*x)

list1=[1,2,3,4,5]
mapped_list = list(map(sqr_num,list1))
print(mapped_list)
````

````python
#this converts string datatype to int we can also use inbuild function as function as well
num=['1','2','3']
map_num= list(map(int,num))
print (map_num)
````

````python
# we can also add multiple list value 


def add(x,y):
  return(x+y)

list1=[1,2,3]
list2=[3,4,5]

map1= list(map(add,list1,list2))
map1
````

**LAMBDA FUNCTION**



*   Anonymus function, which means no name.
*   if we are not going to use specific function more than once rather than creating a function and calling most commonly used function is lambda function.

*   **Syntax: lambda variable_name : return value of function**






````python
# since im not directing using lamba storing the function in an var

x= lambda a:a+10
x(20)
````

````python
#passing more than one value

x = lambda a,b:a+b
x(3,4)
````

````python
#instead creating new function we can use lambda function saves execution time and LOC
lst=[1,2,3]
map_lst= list(map(lambda a:a*a,lst))
map_lst
````

````python
# using multiple list as input using lambda function

n1=[1,2,3]
n2=[3,4,5]
res = list(map(lambda x,y:x+y,n1,n2))
res
````

````python
# to access the value in sheet

for i in range(0, row_count):
    for j in range(0, col_count):
        print(sheet.cell_value(i,j),'\t\t',end='')
    print('\n-----------|')
````

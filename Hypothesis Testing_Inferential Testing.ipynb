# T- Test Two column

Logic is same as Z test 

For T Test - 1 Sample -> we can use Z test to test 
it will give the same output

But for T Test 2 column we will use different code
ttest_ind = ind means independence (Test for independence)

ttest(value 1 , Value2, equal_var = True)

True -> independent H0


```python
import numpy as np
 
group1 = np.array([14, 15, 15, 16, 13, 8, 14, 17, 16, 14, 19, 20, 21, 15, 15, 16, 16, 13, 14, 12])
group2 = np.array([15, 17, 14, 17, 14, 8, 12, 19, 19, 14, 17, 22, 24, 16, 13, 16, 13, 18, 15, 13])
cityA = [82, 84, 85, 89, 91, 91, 92, 94, 99, 99,
        105, 109, 109, 109, 110, 112, 112, 113, 114, 114]
 
cityB = [90, 91, 91, 91, 95, 95, 99, 99, 108, 109,
        109, 114, 115, 116, 117, 117, 128, 129, 130, 133]
 
#ttest_ind(a, b, equal_var=True)
 
#a: an array of sample observations for group 1

#b: an array of sample observations for group 2
#equal_var: if True, perform a standard independent 2 sample t-test that assumes equal population variances. If False, perform Welchs t-test, which does not assume equal population variances. This is True by default.
#Before we perform the test, we need to decide if well assume the two populations have equal variances or not.
#As a rule of thumb, we can assume the populations have equal variances if the ratio of the larger sample variance to the smaller sample variance is less than 4:1.
#find variance for each group
print(np.var(group1), np.var(group2))
 
#The ratio of the larger sample variance to the smaller sample variance is 12.26 / 7.73 = 1.586  i.e<4 so equal population variance
import scipy.stats as stats
 
#perform two sample t-test with equal variances
stats.ttest_ind(a=cityA, b=cityB, equal_var=True)
```

# T TEST MATCHED OR PAIRED SAMPLE TEST

DIFFERENCE BTWN TWO COLUMN TEST AND PARIED SAMPLE TEST:

1. Two column Test: Dependance 

2. Paired Sample Test: Relationship btwn Two column

if they have relationship means they are dependent
and if they have no relation they are independent.

we should consider the same data before and after change for the Matched and paired test 

**Sample for Matched or paired test**

in some cases if we need Result of medicine

consider medicine X

case : we have to find the effect on BP/ engery / insulin, if we use this X medicine 

So inorder to find if there is any effect,

column 1 : We will consider BP before giving the medicine

column 2: We will consider BP after giving the medicine

by doing test on both the column we can find if there are **related to each other or not**.

```python
pre = [88, 82, 84, 93, 75, 78, 84, 87, 95, 91, 83, 89, 77, 68, 91]
post = [91, 84, 88, 90, 79, 80, 88, 90, 90, 96, 88, 89, 81, 74, 92]
 
#bttest_rel(a, b)
 
#where:
 
#a: an array of sample observations from group 1
#b: an array of sample observations from group 2
import scipy.stats as stats
 
#perform the paired samples t-test
stats.ttest_rel(pre, post) # default value H0 (considering it as independent)

'''
p <0.05 --> Reject H0 --> dependant --> we dont dont know if its positive or negative effect 

'''
```

```python
# in this code we are trying to find if the gender affects the salary amt

#Stating of Hypothesis :
#H0 : mu_male = mu_female
#H1 : mu_male != mu_female
male = df[df['gender'] == 'MALE']['income']
female = df[df['gender'] == 'FEMALE']['income']
import scipy.stats as st
st.ttest_ind(male,female) # PROPER SYNTAX: st.ttest_ind(male,female,equal_var = True) by default it will consider as H0 --> independent.
#st.ttest_ind(df["income"],df["age"])

```

if p >0.05 --> H0 --> Failed to Reject --> independent.

```python
# in this code we are trying to find if the with car and without car affects the salary amt

#Stating of Hypothesis :
#H0 : mu_male = mu_female
#H1 : mu_male != mu_female
with_car = df[df['car'] == 'NO']['income']
without_car = df[df['car'] == 'YES']['income']
import scipy.stats as st
st.ttest_ind(with_car,without_car) 
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
 
var=sns.load_dataset("penguins")
#print(var)
 
island=sns.lineplot(x="flipper_length_mm",y="body_mass_g",
hue="island", data=var)

display(island)

species=sns.lineplot(x="flipper_length_mm",y="body_mass_g",
hue="species", data=var)

display(species)

# sns.scatterplot(x="flipper_length_mm",y="body_mass_g",
# hue="island", data=var)
```

```python
#histogram

sns.displot(var, x = 'species',col='island',row='sex', binwidth= 4,hue='species')
```

```python
import seaborn as sns
 
sns.set_theme(style="white",palette='pastel')
df = sns.load_dataset("penguins")
sns.displot(
   df, x="flipper_length_mm", col="species", row="sex",
   binwidth=3, height=4
)

#by default it will only give count 
#if we want to use aggregate we have to aggregate before plot.
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")
 
# Load the example tips dataset
tips = sns.load_dataset("tips")
print(tips)
 
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="sex", y="total_bill",
           hue="smoker", palette=["g", "m"],
           data=tips)
sns.despine(offset=20, left=False,right=False,trim=True) #removes the axis lines 

#dots Exception... if SD is more than 3 it considered as exception

```

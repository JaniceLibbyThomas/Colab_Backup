```python
data = [['guvi',10],['guvi2',12],['guvi3',13]]
df = pd.DataFrame(data,columns=['Name','Age']) #passing the column name
print( df)

# index value by default it takes as 0,1,2... we can also change the index value.
# we can also give the row names, but in real time we wont be using it much...
```

```python
#it converts only the dtype that can only be converted

data = [['guvi',10],['guvi2',12],['guvi3',13]]
df = pd.DataFrame(data,columns=['Name','Age'],dtype=float) #passing the column name
print( df)
```

```python
data = [['guvi',10],['guvi2',12],['guvi3',13]]
df = pd.DataFrame(data,columns=['Name','Age'],index=['a','b','c']) #passing the column name and also row name
print( df)
```

**Uploading the data into dataframe**

```python
import pandas as pd

#Country code 
Country_code = pd.read_csv('https://raw.githubusercontent.com/JaniceLibbyThomas/Sample_Dataset/main/Country-Code.csv')

#Zomato DataSet
Zomato = pd.read_csv('https://raw.githubusercontent.com/JaniceLibbyThomas/Sample_Dataset/main/Zomato.csv')
```

**Converting the Currency to Indian Rupee**

**Exchange rate - INR**

```python
from easy_exchange_rates import API
from datetime import date

Curr_code = list(Country_code['Currency_code'])
Exchange_Rate = []

api = API()
for i in range (0, len(Curr_code)):
  time_series = api.get_exchange_rates(
  base_currency=Curr_code[i], 
  start_date=str(date.today()), 
  end_date=str(date.today()), 
  targets=['INR']
  )
  Exchange_Rate.append((time_series[str(date.today())])['INR'])

Country_code['Exchange_rate'] = Exchange_Rate

Country_code
```

**Merging the Zomato and Country Code**

```python
Zomato = pd.merge(Zomato,Country_code)
Zomato
```

```python
import pandas as pd
data1 = {
   'ID': ['1','2','3','4','5'],
   'Feature1': ['A','B','C','D','E'],
   'Feature2': ['L','M','N',"O",'P']
}
```

```python
df1= pd.DataFrame(data1,columns=['ID','Feature1','Feature2'])
```

```python
df1
```

```python
d2 = {
   'ID': ['1','3','6','7','8'],
   'Feature111': ['AB','BC','CD','DE','EF'],
   'Feature222': ['KL','LM','M-N',"N-O",'O-P']
}
```

```python
df2= pd.DataFrame(d2,columns=['ID','Feature111','Feature222'])
```

```python
df2
```

```python
df3= pd.DataFrame(d2,columns=['ID','Feature11','Feature22'])

df4= pd.DataFrame(d2,columns=['ID1','Feature11','Feature22'])
```

```python
display(df3)

display(df4)
```

```python
#no idential columns

DF2= pd.concat([df1,df4])
display(DF2)

#1 idential columns--> idential column merged and remaining column is added at end....

DF6= pd.concat([df2,df3])
display(DF6)
```

```python
#concat 

#since i have same column name added in row

DF= pd.concat([df1,df2])

display(DF)

#ignores the index of the existing --> after concat follows series no

DF3=pd.concat([df1,df2],ignore_index=True)
display(DF3)

#keys means showing as part of respective table name value

DF4= pd.concat([df1,df2],keys=["1st","2nd"])
display(DF4)

#sorting
DF5= pd.concat([df1,df2],sort=True)
display(DF5)
```

```python
# in a data frame forming a new column by concating col1 and col2

# note: to do both should be same datatype

df1["Feature3"]= df1["Feature1"]+df1["Feature2"]
df1
```

```python
#used to join particular column to an DF
#only unique column will join  
#if there are identical column name it will throw error

Djoin = df1.join(df2["Feature111"]).join(df2["Feature222"])
Djoin
```
